Let us create two series from two dictionaries - one containing car name and price and the other with car name and manufacturer.

````python
#Creating a car price series with a dictionary
car_price_dict = {'Swift':  700000,
                       'Jazz' :  800000,
                       'Civic' : 1600000,
                       'Altis' : 1800000,
                       'Gallardo': 30000000
                      }
car_price = pd.Series(car_price_dict)
# Creating the car manufacturer series with a dictionary
car_man_dict = {'Swift' : 'Maruti',
                  'Jazz'   : 'Honda',
                  'Civic'  : 'Honda',
                  'Altis'  : 'Toyota',
                   'Gallardo' : 'Lamborghini'}
car_man = pd.Series(car_man_dict)
print(car_price)
print(car_man)
````

               columns-  If data contains column labels, it will use the same . Else, default to RangeIndex(0, 1, 2, …, n).

**Problem Statement:**

The teacher does not want to reveal the marks of students who have failed. The condition is that if a student has scored marks >= 33, then they have passed, otherwise failed. The marks of failed students has to be replaced with ‘Fail’. So, how can the task be performed?.

**Solution:**

First is to create the DataFrame as shown below:

````python
marks = [{'Chemistry': 67, 'Physics': 45, 'Mathematics': 50, 'English' : 19},
        {'Chemistry': 90, 'Physics': 92, 'Mathematics': 87, 'English' : 90},
        {'Chemistry': 66, 'Physics': 72, 'Mathematics': 81, 'English' : 72},
        {'Chemistry': 32, 'Physics': 40, 'Mathematics': 12, 'English' : 68}]
marks_df = pd.DataFrame(marks, index = ['Subodh', 'Ram', 'Abdul', 'John'])
marks_df
````

Syntax:

       DataFrame.mask(cond, other = nan, inplace = False, axis = None)

       cond – Where cond is False, keep the original value. Where True, replace with corresponding value from other

       other - Entries where cond is True are replaced with corresponding value from other.

       inplace - Whether to perform the operation in place on the data.

       axis – alignment axis

````python
f_condition = marks_df < 33
marks_df.mask(f_condition, 'Fail')
````

# **Data Manipulation**

In machine learning, a model relies on a dataset for training and testing. However, raw data is often unstructured and not immediately usable. Issues such as missing values (NaN, Null, or NA) and irrelevant rows or columns can hinder the model's performance. To ensure the dataset is clean and optimized for analysis, data wrangling is essential before feeding it into the model.




**Data Manipulation:** The process of organizing, transforming, or restructuring data to make it more suitable for analysis, visualization, or decision-making. This includes filtering, sorting, merging, and modifying data to extract meaningful insights.



**Data Cleaning:** The process of identifying and correcting errors, inconsistencies, or missing values in a dataset to improve its quality and reliability. This includes handling duplicate values, filling or removing missing data, and correcting formatting issues.



**Data Wrangling:** The process of converting raw, unstructured data into a clean and structured format suitable for analysis or machine learning. It involves data cleaning, transformation, and integration to prepare the dataset for further use.
The pandas library in Python is a powerful tool for data manipulation and analysis. The unique function in pandas is used to find the unique values in a Series or DataFrame column. Pandas makes data manipulation easy and efficient with its rich set of functions. The ability to clean, transform, and analyze structured data makes it an essential tool in data science and machine learning.

**value_counts**(): How many times each unique value appears in a column.



**size**: Total number of "cells" in the Data Frame. i.e, it Returns the total number of elements in the Data Frame. This is equivalent to the number of rows multiplied by the number of columns. It includes missing values (NaN)



**count**(): How many valid (non-missing) values are in each column.  Returns the number of non-missing (non-NaN) values in each column (or if applied to the entire Data Frame, in each column). It excludes missing values. It gives you a count of the valid, usable data in each column.
Key Difference

| Method                      | Purpose                                      | Output             | Includes NaN? | Applies to             |
|------------------------------|----------------------------------------------|----------------------|----------------|-------------------------|
| `df['col'].value_counts()`    | Count unique values in a Series              | Series              | No (by default) | Series (single column) |
| `df.size`                    | Total number of elements in DataFrame        | Integer             | Yes             | Data Frame              |
| `df.count()`                 | Count non-missing values in each column      | Series (per column) | No              | Data Frame              |

# **Missing Values**

  They can arise due to various reasons such as data entry errors, incomplete surveys, hardware failures, or system limitations.

  Handling missing values properly is crucial for maintaining the accuracy and reliability of data analysis and machine learning models.

  **Notations Used to Indicate Missing Values**

  Different data sources and programming environments use various notations to represent missing values. The most common ones include:

  **NaN** (Not a Number) – Common in Python libraries like Pandas and NumPy.

  **None** – A Python built-in object used to represent missing values in Pandas.

  **NA** (Not Available) – A common notation in statistical software like R.

  **Blank or Empty Strings** ("") – In text-based data, missing values are sometimes represented as empty cells.

  **Special Symbols** (e.g., ‘?’, ‘-999’, ‘9999’, ‘NA’) – Some datasets use arbitrary values to denote missing data, which need to be identified and handled during preprocessing.


# **Handling Missing Values**

Missing values can be handled in several ways depending on the dataset and the impact on analysis:

1. **Removing missing values**: Deleting rows or columns with missing data if they are not significant.
2. **Imputing missing values**: Replacing missing values with statistical measures such as mean, median, or mode.
3. **Using predictive models**: Estimating missing values based on other available data using machine learning techniques.
Properly addressing missing values ensures that the dataset remains usable and produces accurate insights.

In this section we demonstrate removing and imputing missing values.

````python
import numpy as np
import pandas as pd
# Creating a synthetic dataset with different types of missing values
data = {
    'A': [1, 2, np.nan, 4, 9999, 6, '?', 8, 9],
    'B': [5, '?', 7, 8, 9, np.nan, 11, 9999, 13],
    'C': [9999, 15, 16, '?', 18, np.nan, 20, 21, 22]
}
# Convert the dictionary into a DataFrame
df = pd.DataFrame(data)
# Display original data
print("Original DataFrame:")
print(df)
````

````python
# Replace '?' and 9999 with NaN for standard handling
df.replace(['?', 9999], np.nan, inplace=True)
# Display DataFrame after replacing '?' and 9999 with NaN
print("\nDataFrame after converting special missing values to NaN:")
print(df)
````

````python
# checking missing Value

df.isnull().sum()
````

````python
# Drop missing values
df_dropped = df.dropna()
print("\nDataFrame after dropping missing values:")
print(df_dropped)
````

````python
# Impute missing values with the mean of each column
df_imputed = df.fillna(df.mean(numeric_only=True))
print("\nDataFrame after imputing missing values with column mean:")
print(df_imputed)
````

# **Merge and join in Pandas**

Pandas provides powerful functions like merge() and join() to combine multiple datasets based on common keys or indexes. These functions help in integrating data from different sources for analysis.

**merge() Function**

**Purpose:**

Used to combine two DataFrames based on common columns or indexes.
Similar to SQL JOIN operations.

**Usage:**

When you need to merge on one or more columns.
When combining datasets with different column names.
When performing different types of joins (inner, outer, left, right).

**Syntax :**

pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)

## **Types of merge() Joins:**



| Join Type | Description | Example |
|---|---|---|
| Inner Join | Keeps only matching records from both DataFrames | how='inner' |
| Left Join | Keeps all records from the left DataFrame and matching records from the right | how='left' |
| Right Join | Keeps all records from the right DataFrame and matching records from the left | how='right' |
| Outer Join | Keeps all records from both DataFrames, filling unmatched records with NaN | how='outer' |

## **join() Function**

**Purpose:**

Used to combine two DataFrames based on the index instead of columns.
Efficient when working with indexed data.

**Usage:**

When merging DataFrames based on index values.
When quickly joining hierarchical or time-series data.

**Syntax:**

`df1.join(df2, how='left', on=None)`


**Key Differences Between merge() and join()**

| Feature | merge() | join() |
|---|---|---|
| Merge On | Uses columns as keys | Uses index as keys |
| Flexibility | Can merge on multiple columns with different names | Works best with indexed DataFrames |
| Performance | Slightly slower for large datasets | Faster when using index-based joins |
| Common Use Case | Combining tabular data with matching keys | Joining time-series or hierarchical data |


**When to Use merge() vs join()?**

* **Use merge()** when you are joining on a specific column.
* **Use join()** when you are joining based on the index.
* If merging large datasets, optimize indexing to improve performance.

Both methods are essential for data preprocessing in real-world scenarios like data integration, feature engineering, and database-like operations in Pandas.

Use merge() for column-based joins and join() for index-based joins.

**There are different ways to join the data frames using merg() and join().**

The choice of merge type depends on the specific data you have and what you want to achieve:

* Inner: Use when you only want to keep data that is common to both DataFrames.
* Outer: Use when you want to keep all data from both DataFrames, even if there are no matches.
* Left: Use when you want to keep all data from the "left" DataFrame and matching data from the "right" DataFrame (common when the left DataFrame is your primary source of information).
* Right: Use when you want to keep all data from the "right" DataFrame and matching data from the "left" DataFrame.

Understanding these merge types is essential for effective data manipulation in Pandas. The pd.merge() function is incredibly versatile and is a fundamental tool for data analysis.

***Below set of code demonstrates different types join and merge using a synthetic dataset.***

````python
import pandas as pd
# Sample DataFrames
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'col1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D', 'E'], 'col2': [5, 6, 7, 8]})
print("Data Frame -1 \n", df1)
print("Data Frame -2\n",df2)
````

````python
# 1. Inner Merge
merged_inner = pd.merge(df1, df2, on='key', how='inner')
print("Inner Merge:\n", merged_inner)
````

````python
# 2. Outer Merge
merged_outer = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Merge:\n", merged_outer)
````

````python
# 3. Left Merge
merged_left = pd.merge(df1, df2, on='key', how='left')
print("\nLeft Merge:\n", merged_left)
````

````python
# 4. Right Merge
merged_right = pd.merge(df1, df2, on='key', how='right')
print("\nRight Merge:\n", merged_right)
````

````python
# --- Join Operations (assuming 'key' is the index in both DataFrames) ---
df1_indexed = df1.set_index('key')
df2_indexed = df2.set_index('key')
````

````python
# 1. Inner Join
joined_inner = df1_indexed.join(df2_indexed, how='inner')
print("\nInner Join (using join):\n", joined_inner)
````

````python
# 2. Outer Join
joined_outer = df1_indexed.join(df2_indexed, how='outer')
print("\nOuter Join (using join):\n", joined_outer)
````

````python
# 3. Left Join
joined_left = df1_indexed.join(df2_indexed, how='left')
print("\nLeft Join (using join):\n", joined_left)
````

````python
# 4. Right Join
joined_right = df1_indexed.join(df2_indexed, how='right')
print("\nRight Join (using join):\n", joined_right)
````

# **Melt**

* The pd.melt() function in Pandas is used to reshape a DataFrame from a wide format to a long format. This is also known as "unpivoting." It's particularly useful when you have data where multiple columns represent the same type of measurement or variable, and you want to transform this into a more manageable format for analysis or visualization.

**Analogy:**

* Imagine you have a table of sales data where each column represents a month (e.g., "January," "February," "March") and each row represents a product. The values in the table are the sales figures for each product in each month. pd.melt() can transform this table into a long format where you have two columns: "Month" and "Sales." Each row in the new table would then represent a single product's sales in a specific month.

How pd.melt() works ?

pd.melt() takes several key arguments:

```
frame: The DataFrame you want to melt.
id_vars: A list of column names that you want to keep as identifier variables. These columns will not be "unpivoted." They are used to identify each row.

value_vars: A list of column names that you want to unpivot (melt). These are the columns that represent the same type of measurement. If not provided, all columns except id_vars will be unpivoted.

var_name: The name you want to give to the new column that will contain the names of the unpivoted columns (e.g., "Month" in the sales example).

value_name: The name you want to give to the new column that will contain the values from the unpivoted columns (e.g., "Sales").

```



````python
import pandas as pd
data = {'Product': ['A', 'B', 'C'],
        'January': [100, 150, 200],
        'February': [120, 180, 250],
        'March': [140, 210, 300]}
df_wide = pd.DataFrame(data)
print("Wide Format DataFrame:\n", df_wide)
````

````python
# Another example, melting all columns except 'Product'
df_long_all_vars = pd.melt(df_wide, id_vars=['Product'], var_name='Month', value_name='Sales')
print("\nLong Format DataFrame (all vars):\n", df_long_all_vars)
````

````python
df_long = pd.melt(df_wide,
                  id_vars=['Product'],
                  value_vars=['January', 'February', 'March'],
                  var_name='Month',
                  value_name='Sales')
print("\nLong Format DataFrame:\n", df_long)
````
